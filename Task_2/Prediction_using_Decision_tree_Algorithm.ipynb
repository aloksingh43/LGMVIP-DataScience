# **Let's remove outliers**

```python
q1=df['sepal width (cm)'].quantile(0.25)
q3=df['sepal width (cm)'].quantile(0.75)
print(q1)
print(q3)
```

```python
IQR=q3-q1
Lower_Whisker = q1-1.5*IQR
Upper_Whisker = q3+1.5*IQR
print(Lower_Whisker)
print(Upper_Whisker)
```

```python
df=df[df['sepal width (cm)']>Lower_Whisker]
df=df[df['sepal width (cm)']<Upper_Whisker]
```

---
# **Split Variable**
---

X = Independent Variable

y = Dependent Variable

```python
X= df.drop(columns=['Target'])
y=df['Target']
```

---
# **Train Test Split**
---

Splitting Data

```python
X_train, X_test, y_train, y_test = train_test_split( X,y , test_size = 0.2, random_state = 0) 
```

# **Decision Tree Classifier**

```python
# apply Decision Tree classifier
dtr_model = DecisionTreeClassifier()
dtr_model.fit(X_train, y_train)
```

```python
pred_test = dtr_model.predict(X_test)
```

```python
# print classification report
clf_report= classification_report(y_test, pred_test)
```

```python
print(clf_report)
```

```python
print(f"Accuracy Score: {accuracy_score(y_test, pred_test) * 100:.2f}%")
```

```python
# print confusion matrix
cm_test=confusion_matrix(y_test, pred_test)
sns.heatmap(cm_test, annot=True, fmt='d');
```

```python
# print decision tree
graph = Source(tree.export_graphviz(dtr_model, out_file=None
   , feature_names=X_train.columns, class_names=['0', '1','2'] 
   , filled = True))
display(SVG(graph.pipe(format='svg')))
```

**Hyperparatmeter Tunning**

```python
dtr_model_tuning = DecisionTreeClassifier()
parameters_decision_tree = { "max_depth": [None, 4,9,15,20,50],
    "min_samples_leaf": [ 1,4,0.1,2,10],
    "max_features" : [None,'auto', 'sqrt', 'log2'],
    "min_samples_split": [2,9,15,25]}
```

**max_depth** - The maximum depth of the tree. If None, then nodes are expanded until all leaves are pure or until all leaves contain less than min_samples_split samples.

**min_samples_leaf** - The minimum number of samples required to be at a leaf node. A split point at any depth will only be considered if it leaves at least min_samples_leaf training samples in each of the left and right branches.

**max_features**-The number of features to consider when looking for the best split.

**min_samples_split**-The minimum number of samples required to split an internal node.


```python
#  apply GridSearchCV
tuned_mode = GridSearchCV(dtr_model_tuning, param_grid = parameters_decision_tree, scoring ='accuracy', cv = 5)
tuned_mode.fit(X_train, y_train)
```

```python
tuned_mode.best_score_
```

```python
print(tuned_mode.best_estimator_)
```

```python
pred_test_tuned = tuned_mode.predict(X_test)
```

```python
print(f"Accuracy Score: {accuracy_score(y_test, pred_test_tuned) * 100:.2f}%")
```

```python
clf_report_tuned= classification_report(y_test, pred_test_tuned)
print(clf_report_tuned)
```

```python
cm_test=confusion_matrix(y_test, pred_test_tuned)
sns.heatmap(cm_test, annot=True, fmt='d');
```
