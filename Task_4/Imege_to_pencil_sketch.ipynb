### **Loading Image**


**Display the Image**

```python
img = cv2.imread('/content/drive/MyDrive/letsgrow/GettyImages-988013222-scaled-e1618857975729.jpg')
cv2_imshow(img)
```

```python
# change into gray
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
```

```python
# invert gray imege
invert=cv2.bitwise_not(img_gray)
cv2_imshow(invert)
```

```python
# make imege blur
img_blur = cv2.GaussianBlur(invert, (21,21), 0, 0)
cv2_imshow(img_blur)
```

```python
# invert blur imege
inverted_blur = cv2.bitwise_not(img_blur)
cv2_imshow(inverted_blur)
```

```python
#combine all and make skatch imege
sketch=cv2.divide(img_gray,inverted_blur,scale=240)
print('sketch imege')
cv2_imshow(sketch)
```
